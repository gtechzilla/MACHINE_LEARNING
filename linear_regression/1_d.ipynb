```python
#importing the necessary libraries

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#loading the data
data=pd.read_csv('data_1d.csv',header=None)

#plotting the data and converting to numpy arrays
X=np.array(data.iloc[:,0])
Y=np.array(data.iloc[:,1])
plt.scatter(X,Y)

```

```python
#denominator is common for a and b
denominator=np.mean(X.dot(X))-(np.mean(X))**2
print(denominator)
```

```python
#working out the values for a and b
a=((np.mean(np.dot(X,Y)))-(np.mean(X)*np.mean(Y)))/denominator
print(a)
b=(((np.mean(np.dot(X,X))*np.mean(Y)))-((np.mean(X)*np.mean(np.dot(X,Y)))))/denominator
print(b)

#working out value of Yhat
Yhat= a*X + b
print(Yhat)
```

```python
#plotting line of best fit

fig=plt.figure() #creates a canvas object

ax = fig.add_axes([0,0,1,1]) #sets location of graph on the canvas

ax.scatter(X,Y,label='data set') #plots a scatter plot for the data
ax.plot(X,Yhat,'g-',label='prediction') #plots line of best fit

ax.legend(loc=0) #sets location of legend to top left

ax.set_title("1 Dimension Linear Regression")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
```

```python
#solving for R squared
numerator_R=Y-Yhat
denominator_R=Y-Y.mean()

#we use the dot function as bot the numerator and denominator are squared
r2=1-numerator_R.dot(numerator_R)/denominator_R.dot(denominator_R)
print('the r-squared is ',r2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data=pd.read_csv('tvmarketing.csv')
```

```python
train_data.head()
```

```python
#getting info about our columns
train_data.info()
```

```python
#getting statistical info about our data columns
train_data.describe()
```

```python
#shape of our data
train_data.shape
```

```python
#visualizing the data using seaborn

#a pairplot is used to show how two features relate with one another
import seaborn as sns
sns.pairplot(train_data,vars=['TV','Sales'])
```

```python
#plotting a single pairplot of sales against tv
sns.pairplot(train_data, x_vars=['TV'], y_vars='Sales',height=5, aspect=0.7, kind='scatter')
```

```python
#separating our data into features and target
#features are in the X axis while targets in the Y
X=train_data['TV']
Y=train_data['Sales']
```

```python
#splitting the data into test and train datasets
#we use the train_test split function fromsklearns model_selection module
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7,random_state=42)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
#from above we that our data only has rows but no columns
#with scikit learn we have to have feature,which are represented as columns while observations are recorded as rows
#to ammend this we use numpys newaxis method,this adds a column to our array
X_train= X_train[:,np.newaxis]
Y_train=Y_train[:,np.newaxis]
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
#importing our model
from sklearn.linear_model import LinearRegression
#creating an object to store our model
lr=LinearRegression()

#fitiing our model to the train_data
lr.fit(X_train,Y_train)
```

```python
#checking for the y intercept and coefficient of our line equation
print(lr.coef_)
print(lr.intercept_)
```

```python
X_test=X_test[:,np.newaxis]
#making our predictions on our test split data
Y_predicted=lr.predict(X_test)
```

```python
#plotting our predicted sales values
c=[i for i in range(1,61,1)]
fig=plt.figure()
plt.plot(c,Y_test,color='blue',linewidth="2.5",linestyle="-")
plt.plot(c,Y_predicted,color='red',linewidth="2.5",linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                               # X-label
plt.ylabel('Sales', fontsize=16) 
```

```python
Y_test=Y_test[:,np.newaxis]
```

```python
c=[i for i in range(1,61,1)]
fig=plt.figure()
plt.plot(c,Y_test-Y_predicted,color="green",linewidth="2.5",linestyle="-")
fig.suptitle('our error',fontsize=20)
"""plt.xlabel=('Index', fontsize=18)
plt.ylabel=('Error', fontsize=16)"""
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(Y_test,Y_predicted)
```

```python
r2=r2_score(Y_test,Y_predicted)
```

```python
print('Mean Squared error is',mse)
print('r squared error is',r2)
```
